```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
dataset=pd.read_csv("HR-Employee-Attrition.csv")
''' This dataset tells us about the attrition of employees based on various situations.We want to predict whether the employee 
    will leave or not. So our goal is to find optimized value of C and gamma '''
```

```python
dataset
```

```python
X=dataset.iloc[:,[0,3,5,6]]
```

```python
X
```

```python
y=dataset.iloc[:,1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=41)
```

```python
from sklearn.svm import SVC
model=SVC()
```

```python
model.fit(X_train,y_train)
```

```python
predictions=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
print('\n')
print(classification_report(y_test,predictions))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'C' :[0.1,1,10,100,1000],'gamma':[1,0.1,0.01,0.001,0.0001]}
```

```python
grid=GridSearchCV(SVC(),param_grid,verbose=3)
#Verbose is included to show line by line compilation of output. One may choose any value. The higher the value the more is verbose
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
#This command shows the optimized values of C(cost of miscalculations) and gamma
```

```python
grid.best_estimator_
```

```python
grid_predictions=grid.predict(X_test)
```

```python
print(confusion_matrix(y_test,grid_predictions))
print('\n')
print(classification_report(y_test,grid_predictions))
# Now one may see the confusion matrix where the 0 values are not there and accuracy has also improved
```
